# Fake Agent UI Mock

The actual app is a Streamlit script and lives next to this notebook:
[`fake_agent_app.py`](./fake_agent_app.py). Streamlit cannot be hosted from
inside a Jupyter notebook, so this notebook is just a placeholder with the
launch instructions.

## Start

From the repository root, run:

```bash
make streamlit-run
```

Defaults: address `0.0.0.0`, port `8501`, file
`assorted/session3/fake_agent_app.py` (relative to `/workspace/src` inside
the dev container). Override per call:

```bash
make streamlit-run STREAMLIT_PORT=8502
make streamlit-run STREAMLIT_FILE=assorted/session3/other_app.py
```

The dev container currently does not publish port `8501` to the host. To
reach the UI from your browser, either add a port mapping for the `dev`
service in `container/compose/docker-compose.yml`, or use `docker compose
port dev 8501` / SSH port forwarding.

## Scope

UI only. The Chaos slider is interactive but triggers no action. The
prompt textarea and Send button do nothing on input/click. The "Thinking"
panel shows four placeholder agents whose text will be filled in later.
